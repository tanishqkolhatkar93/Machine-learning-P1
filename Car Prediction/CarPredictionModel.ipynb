* THE FEATURES, 'Fuel_Type', 'Seller_Type' , 'Transmission' are CATEGORICAL FEATURES. 
* Categorical Features are those which don't take a mathematical number as inputs, so we need to pre process them
* before entering them into the model.

* We do that by using DUMMY FEATURES. FUNCTION: GET DUMMY

```python
print(df['Fuel_Type'].unique())
```

* Now since , our categorical features are very less, we will be *ONE- HOT ENCODING* them.
* SINCE THE CAR COULD BE EITHER Petrol, Diesel or CNG and NOT ANY TWO AT THE SAME TIME.
* SO, 
*  PETROL     DIESEL    CNG
*  1               0         0
*  0               1         0
*  0               0         1

* if it is petrol type, then it wont be diesel or cng type and similarly for all the three cases.

```python
final_dataset= pd.get_dummies(final_dataset)     #GET DUMMIES, MAKES THEM ONE HOT ENCODED
```

```python
final_dataset.head()
```

**NOW, IF WE DROP THE FIRST FEATURE, WE WONT INCURR ANY LOSS, BECAUSE IF WE HAVE THE OTHER TWO VALUES WE CAN  AUTOMATICALLY COUNT THE 3RD ONE. 
**# SO, IN ORDER TO ESCAPE THE *DUMMY VARIABLE TRAP*,  WE WILL DROP THE FIRST FEATURE OUT OF THE THREE, 'FUEL_TYPE_CNG'.
**The solution to this problem is achieved by incorporating the concept of dummy columns. For each of the values of a certain category, a new column is introduced. 

```python
final_dataset= pd.get_dummies(final_dataset,drop_first= True)  
```

*Random Forest Regression is a supervised learning algorithm that uses ensemble learning method for regression.
A Random Forest operates by constructing several decision trees during training time and outputting the mean of the 
classes as the prediction of all the trees.

This situation where any given model is performing too well on the training data but the performance drops significantly over the test set is called an *overfitting* model. On the other hand, if the model is performing poorly over the test and the train set, then we call that an *underfitting* model.

In machine learning, *hyperparameter optimization or tuning* is the problem of choosing a set of optimal hyperparameters for a learning algorithm. A hyperparameter is a parameter whose value is used to control the learning process. 


```python
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
import numpy as np
```

```python
###HYPERPARAMETERS
###selecting different decision tree values
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
print(n_estimators)
```

```python
from sklearn.model_selection import RandomizedSearchCV     #RandomizedSearchCV does hyperparameter tuning
#Randomized Search CV
#initialising parameters

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
# using them in the format of key-value pairs
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)                    #RandomizedSearchCV will choose the best parameters of the given ones.
```

```python
# USE THE RANDOM GRID TO SEARCH FOR BEST HYPERPARAMETERS
# FIRST CREATE THE BASE MODEL TO TUNE.
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(x_train, y_train)
```

```python
predictions = rf_random.predict(x_test)
```
